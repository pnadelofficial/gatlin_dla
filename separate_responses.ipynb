# Pre

```python
import pandas as pd
from DLA import DLAnalyzer
```

```python
consol_pre = pd.read_csv('./data/pre_consolidated.csv')
sep_pre = consol_pre[['ID'] + list(consol_pre.columns[14:-2])]
```

```python
sep_pre = sep_pre.rename(columns={
    '0':'What underlying assumptions, attitudes or beliefs do you have of canine breeders and the canine breeder community?',
    '1':'What underlying assumptions, attitudes or beliefs do you have of purebred and purpose bred dogs?',
    '2':'What underlying assumptions, attitudes or beliefs do you have of canine shows (conformation, agility etc.)?',
    '3':'What topics and/or careers in small anilam theriogenology and canine genetics that are you interested in exploring in the selective?',
    '4':'What resources can we at Tufts provide you with, from a veterinary perspective?',
    '5':'What resources can your breeder mentor or AKC provide you with, from a breeder community perspective?'
})
```

```python
sep_pre
```

## Without last question

```python
sep_pre1 = sep_pre[:-1]
sep_pre1
```

```python
sep_pre1[sep_pre1.columns[1+5]]
```

```python
likert_responses = []
for i in range(1,6):
    df = pd.DataFrame([sep_pre1[sep_pre1.columns[0]], sep_pre1[sep_pre1.columns[i]], sep_pre1[sep_pre1.columns[i+5]]]).T
    likert_responses.append(df)
```

```python
likert_responses[0]
```

```python
for df in likert_responses:
    out = df.columns[1]
    res = df.columns[2]
    print(out, res, sep='\n')
    print('-'*20)
    dla = DLAnalyzer(df, res, out, 'ID', alpha=.1, min_narr_size=0)
    dla.dla()
    dla.apply_threshold()
    print(dla.pos, dla.neg)
```

## With last question

```python
sep_pre2 = sep_pre
sep_pre2
```

```python
likert_responses2 = []
for i in range(1,6):
    df = pd.DataFrame([sep_pre2[sep_pre1.columns[0]], sep_pre2[sep_pre1.columns[i]], sep_pre2[sep_pre1.columns[i+5]], sep_pre2[sep_pre1.columns[-1]]]).T
    df['combined_response'] = df[df.columns[2]] + df[df.columns[3]]
    likert_responses2.append(df)
```

```python
likert_responses2[0]
```

```python
for df in likert_responses2:
    out = df.columns[1]
    que = df.columns[2]
    res = df.columns[-1]
    print(out, que, sep='\n')
    dla = DLAnalyzer(df, res, out, 'ID', alpha=.1, min_narr_size=0)
    dla.dla()
    dla.generate_word_clouds()
    # except:
    #     dla.apply_threshold()
    #     print(dla.pos, dla.neg)
    print('-'*20)
```

# Post

```python
import pandas as pd
from DLA import DLAnalyzer
```

```python
consol_post = pd.read_csv('./data/post_consolidated.csv')
sep_post = consol_post[['ID'] + list(consol_post.columns[14:-2])]
```

```python
sep_post = sep_post.rename(columns={
    '0':'What underlying assumptions, attitudes or beliefs do you have of canine breeders and the canine breeder community?',
    '1':'What underlying assumptions, attitudes or beliefs do you have of purebred and purpose bred dogs?',
    '2':'What underlying assumptions, attitudes or beliefs do you have of canine shows (conformation, agility etc.)?',
    '3':'What topics and/or careers in small anilam theriogenology and canine genetics that are you interested in exploring in the selective?',
    '4':'What resources can we at Tufts provide you with, from a veterinary perspective?',
    '5':'What resources can your breeder mentor or AKC provide you with, from a breeder community perspective?'
})
```

```python
sep_post
```

## Without last question

```python
sep_post1 = sep_post[sep_post.columns[:-2]]
sep_post1
```

```python
likert_responses = []
for i in range(1,6):
    df = pd.DataFrame([sep_post1[sep_post1.columns[0]], sep_post1[sep_post1.columns[i]], sep_post1[sep_post1.columns[i+6]]]).T
    likert_responses.append(df)
```

```python
likert_responses[0]
```

```python
for df in likert_responses:
    out = df.columns[1]
    que = df.columns[2]
    res = df.columns[-1]
    print(out, que, sep='\n')
    dla = DLAnalyzer(df, res, out, 'ID', alpha=.05, min_narr_size=0)
    dla.dla()
    dla.generate_word_clouds()
    dla.regress()
    print(dla.regscore)
    print('-'*20)
```
