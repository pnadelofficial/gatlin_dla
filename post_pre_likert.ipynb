```python
import pandas as pd
from DLA import DLAnalyzer
```

```python
consol_pre = pd.read_csv('./data/pre_consolidated.csv')
sep_pre = consol_pre[['ID'] + list(consol_pre.columns[14:-2])]
sep_pre_col_names = ['ID', 'Q14', 'Q15', 'Q16', 'Q18', 'Q19', 'Q20', 'Q21', 'Q22', 'Q23', 'Q24', 'Q25']
sep_pre.columns = sep_pre_col_names
sep_pre
```

```python
consol_post = pd.read_csv('./data/post_consolidated.csv')
sep_post = consol_post[['ID'] + list(consol_post.columns[14:-3])]
sep_post_col_names = ['ID', 'Q15', 'Q16', 'Q17', 'Q18', 'Q19', 'Q20', 'Q21', 'Q22', 'Q23', 'Q24', 'Q25', 'Q26']
sep_post.columns = sep_post_col_names
sep_post
```

```python
fr_columns = ['Q21', 'Q22', 'Q23', 'Q24', 'Q25', 'Q26']
def likert_comparison(post_q, pre_q, fr_columns=fr_columns, wc=True, **kwargs):
    outs = sep_post[post_q] - sep_pre[pre_q]
    fr_concat  = sep_post.apply(lambda row: " ".join(row.loc[fr_columns]), axis=1)
    df = pd.DataFrame({'ID': sep_post['ID'], 'likert_diff': outs, 'free_response_concatenated':fr_concat})
    dla = DLAnalyzer(df, 'free_response_concatenated', 'likert_diff', 'ID', name=f"likert_comparison{post_q}_{pre_q}", **kwargs)
    dla.dla()
    if wc: dla.generate_word_clouds(path='../../Library/CloudStorage/Box-Box/Canine Repro Selective_PN_MSG/word clouds', use_name=True)
    return dla
```

```python
likert_comparison('Q15', 'Q14', lemmatize=True)
```
