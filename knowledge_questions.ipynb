```python
import pandas as pd
from DLA import DLAnalyzer
```

```python
pre = pd.read_csv('./data/pre.csv')
pre
```

```python
post = pd.read_csv('./data/post.csv')
post
```

```python
qs = dict(zip(list(pre.columns[1:13]), [f'Q{i}' for i in range(1, 13)]))
qs
```

```python
pre = pre.rename(columns=qs)
post = post.rename(columns=qs)
```

```python
outs = post['Q1'] - pre['Q1']
```

```python
consol_post = pd.read_csv('./data/post_consolidated.csv')
sep_post = consol_post[['ID'] + list(consol_post.columns[14:-3])]
sep_post_col_names = ['ID', 'Q15', 'Q16', 'Q17', 'Q18', 'Q19', 'Q20', 'Q21', 'Q22', 'Q23', 'Q24', 'Q25', 'Q26']
sep_post.columns = sep_post_col_names
sep_post
```

```python
fr_columns = ['Q21', 'Q22', 'Q23', 'Q24', 'Q25', 'Q26']
def knowledge_comparison(knowledge_question, fr_columns, **kwargs):
    outs = post[knowledge_question] - pre[knowledge_question]
    fr_concat  = sep_post.apply(lambda row: " ".join(row.loc[fr_columns]), axis=1)
    df = pd.DataFrame({'ID': sep_post['ID'], 'likert_sum': outs, 'free_response_concatenated':fr_concat})
    dla = DLAnalyzer(df, 'free_response_concatenated', 'likert_sum', 'ID', **kwargs)
    dla.dla()
    dla.generate_word_clouds()
    return dla
```

```python
knowledge_comparison('Q1', fr_columns, alpha=.05, min_word_freq=0, remove_stops=True, lemmatize=False)
```

```python
dla = knowledge_comparison('Q2', fr_columns, alpha=.05, min_word_freq=1, remove_stops=True, lemmatize=True)
```

```python
dla.search('profit', nb=True)
```
