# data prep

```python
import os
base_path = '../../Library/CloudStorage/Box-Box/Canine Repro Selective_PN_MSG'
os.listdir(base_path)
```

```python
import pandas as pd

pre_df =  pd.read_csv(f'{base_path}/Pre_Data_PN_MSG.csv', skiprows=[0])
pos_df = pd.read_csv(f'{base_path}/Post_Data_PN_MSG.csv', skiprows=[0])
```

```python
pre_df = pre_df.reset_index(drop=True).iloc[:7]
pre_df['ID'] = 'Student_' + pre_df.Student.astype(int).astype(str)
pre_df[pre_df.columns[1:-1]] = pre_df[pre_df.columns[1:-1]].astype(float)
```

```python
pre_df = pre_df[[pre_df.columns[-1]] + list(pre_df.columns[1:-1])]
pre_df
```

```python
def format_df(path):
    df = pd.read_csv(path, skiprows=[0])
    df = df.reset_index(drop=True).iloc[:7]
    df['ID'] = 'Student_' + df.Student.astype(int).astype(str)
    df[df.columns[1:-1]] = df[df.columns[1:-1]].astype(float)
    df = df[[df.columns[-1]] + list(df.columns[1:-1])]
    return df
```

```python
pre_df  = format_df(f'{base_path}/Pre_Data_PN_MSG.csv')
post_df = format_df(f'{base_path}/Post_Data_PN_MSG.csv')
```

```python
pre_df.to_csv('./data/pre.csv', index=False)
post_df.to_csv('./data/post.csv', index=False)
```

```python
pre_qual_df =  pd.read_csv(f'{base_path}/Pre Qual.csv')
pos_qual_df = pd.read_csv(f'{base_path}/Post Qual.csv')
```

```python
pre_qual_df.T[1:][[0,1,2,3,4,5]].reset_index()
```

```python
pos_qual_df.T[1:].reset_index(drop=True)
```

```python
pre = pre_df.merge(pre_qual_df.T[1:][[0,1,2,3,4,5]].reset_index(drop=True), left_index=True, right_index=True)
post= pos_df.merge(pos_qual_df.T[1:].reset_index(drop=True), left_index=True, right_index=True)
```

```python
pre.columns[13:-6]
```

```python
likert_cols = pre.columns[13:-6].to_list()
free_res_cols = pre.columns[-6:]
free_res_cols
```

```python
consol_pre = pre.copy()
consol_pre['consol_sum'] = consol_pre[likert_cols].sum(axis=1)
consol_pre['consol_res'] = consol_pre[free_res_cols].apply(lambda x: '\n'.join(x), axis=1)
```

```python
consol_pre
```

```python
consol_pre.to_csv('./data/pre_consolidated.csv')
```

```python
likert_cols = post.columns[13:-6].to_list()
free_res_cols = post.columns[-6:]

consol_post = post.copy()
consol_post['consol_sum'] = consol_post[likert_cols].sum(axis=1)
consol_post['consol_res'] = consol_post[free_res_cols].apply(lambda x: '\n'.join(x), axis=1)
```

```python
consol_post['ID'] = consol_post['Student'].apply(lambda x: 'Student_'+str(int(x)))
```

```python
consol_post
```

```python
consol_post.to_csv('./data/post_consolidated.csv')
```
