```python
import pandas as pd
from DLA import DLAnalyzer
```

```python
qs = {
    'Q14':'There is a healthy relationship between breeders and vets',
    'Q15':'There is a community need for purebred dogs',
    'Q16':'There is adequate education material in Tufts preclinical curriculum',
    'Q17':'There is adequate education material in Tufts clinical curriculum',	 
    'Q18':'After graduation, I am likely to practice with canine breeder clientele',	 
    'Q19':'I feel confident graduating from Tufts and practicing with canine breeder clientele',
    'Q20':'What underlying assumptions, attitudes or beliefs do you have of canine breeders and the canine breeder community?',
    'Q21':'What underlying assumptions, attitudes or beliefs do you have of purebred and purpose bred dogs?',
    'Q22':'What underlying assumptions, attitudes or beliefs do you have of canine shows (conformation, agility etc.)?',
    'Q23':'What topics and/or careers in small animal theriogenology and canine genetics that are you interested in exploring in the selective?',
    'Q24':'What resources can we at Tufts provide you with, from a veterinary perspective?',
    'Q25':'What resources can your breeder mentor or AKC provide you with, from a breeder community perspective?'
}
```

```python
def compare_likert_fr(org_df, likert_cols, fr_columns, alpha=.05):
    likert_sum = org_df[likert_cols].sum(axis=1)
    fr_concat  = org_df.apply(lambda row: " ".join(row.loc[fr_columns]), axis=1)
    df = pd.DataFrame({'ID': org_df['ID'], 'likert_sum': likert_sum, 'free_response_concatenated':fr_concat})
    dla = DLAnalyzer(df, 'free_response_concatenated', 'likert_sum', 'ID', alpha=alpha, min_narr_size=0, min_word_freq=2)
    dla.dla()
    # dla.apply_threshold()
    # print(dla.pos)
    # print(dla.neg)
    dla.generate_word_clouds()
    # dla.regress() # report r-squared
```

# Pre

```python
consol_pre = pd.read_csv('./data/pre_consolidated.csv')
sep_pre = consol_pre[['ID'] + list(consol_pre.columns[14:-2])]
sep_pre_col_names = ['ID', 'Q14', 'Q15', 'Q16', 'Q18', 'Q19', 'Q20', 'Q21', 'Q22', 'Q23', 'Q24', 'Q25']
sep_pre.columns = sep_pre_col_names
sep_pre
```
